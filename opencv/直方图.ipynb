```python
import cv2
import numpy as np
def cv2_show(im):
    cv2.namedWindow('image',cv2.WINDOW_NORMAL)
    cv2.imshow('image',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
im=cv2.imread('../data/video/cat.jpg')
hist_b=cv2.calcHist([im],[0],None,[256],[0,256])
hist_g=cv2.calcHist([im],[1],None,[256],[0,256])
hist_r=cv2.calcHist([im],[2],None,[256],[0,256])
```

```python
import matplotlib.pyplot as plt
plt.hist(im.ravel(),256)
```

```python
plt.plot(hist_b,color='b')
plt.plot(hist_g,color='g')
plt.plot(hist_r,color='r')
plt.show()
```

```python
mask=np.zeros(im.shape[:2],np.uint8)
mask[100:300,100:300]=255
cv2_show(mask)
```

```python
hist_mask=cv2.calcHist([im],[0],mask,[256],[0,256])
plt.plot(hist_mask,color='b')
plt.plot(hist_b,color='g')
```
