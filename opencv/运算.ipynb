```python
import cv2
def cv2_show(im):
    cv2.namedWindow('image',cv2.WINDOW_NORMAL)
    cv2.imshow('image',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
top,bottom,left,right=(50,50,50,50)
replicate=cv2.copyMakeBorder(cat,top,bottom,left,right,cv2.BORDER_REPLICATE)
reflect=cv2.copyMakeBorder(cat,top,bottom,left,right,cv2.BORDER_REFLECT)
reflect101=cv2.copyMakeBorder(cat,top,bottom,left,right,cv2.BORDER_REFLECT_101)
wrap=cv2.copyMakeBorder(cat,top,bottom,left,right,cv2.BORDER_WRAP)
constant=cv2.copyMakeBorder(cat,top,bottom,left,right,cv2.BORDER_CONSTANT,value=(0,0,0))
operation=[cat,reflect,reflect101,reflect,wrap,constant]
title=['cat','replicate','reflect','reflect101','wrap','constant']
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(2,3)
    for i in range(2):
        for j in range(3):
            ax[i,j].imshow(operation[j+3*i])
            ax[i,j].axis('off')
            ax[i,j].set_title(title[j+3*i])
    
```
