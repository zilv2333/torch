```python
import cv2
def cv2_show(im):
    cv2.namedWindow('image',cv2.WINDOW_NORMAL)
    cv2.imshow('image',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```


ret,dst=cv2.threshold(src,thresh,maxval,type)
src：输入单通道
dst：输出值
thresh：阈值
maxval:超过(低于）阈值所赋的值（取决于type)
type:cv2.THRESH_BINARY 超过阈值取maxval，否则0
cv2.THRESH_BINARY_INV 反转
cv2.THRESH_TRUNC 大于阈值部分取阈值
cv2.THRESH_TOZERO 大于部分不变，小于取0
cv2.THRESH_TOZERO_INV 反转

```python
im=cv2.imread('../data/video/cat.jpg')
types=[cv2.THRESH_BINARY,
       cv2.THRESH_BINARY_INV ,
       cv2.THRESH_TRUNC,
       cv2.THRESH_TOZERO,
       cv2.THRESH_TOZERO_INV ]
title=['THRESH_BINARY',
       'THRESH_BINARY_INV' ,
       'THRESH_TRUNC',
       'THRESH_TOZERO',
       'THRESH_TOZERO_INV' ]
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,6,figsize=(15,5))
for j in range(5):
    ret,thresh=cv2.threshold(im,127,255,types[j])
    ax[j].imshow(thresh)
    ax[j].axis('off')
    ax[j].set_title(title[j])
ax[5].axis('off')
ax[5].imshow(im)
ax[5].set_title('cat')
```
